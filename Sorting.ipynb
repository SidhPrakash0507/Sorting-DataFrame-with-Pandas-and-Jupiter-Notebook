```python
import pandas as pd
data=pd.read_excel("C:\\Users\\Data\\Data for Sorting in Pandas Lib.xlsx")#Giving path to read Data of Excel
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df.sort_values("Maths",ascending=True)#Now Here it Prints the Value in ascending order if you want to put it in desending then change the attribute value to false given in below cell
```

```python
df.sort_values("Maths",ascending=False)#Now here it Print the value in decending order
```
